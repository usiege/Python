```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import animation

# 设置图表风格
sns.set_style("whitegrid")

# 生成示例数据
data = sns.load_dataset("tips") 

# 创建图表对象
fig, ax = plt.subplots()

# 绘制散点图
scatter = sns.scatterplot(
    x="total_bill", y="tip", 
    data=data, ax=ax
)

# 定义更新函数
def update(i):
    scatter.collections[0].set_offsets(data.sample(n=10).values)

# 生成动画    
ani = animation.FuncAnimation(fig, update, frames=100, interval=200)

# 保存为gif
ani.save('scatter.gif', writer='imagemagick')

plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import random

# 输入文本数据
text = "清洁用品 爆炸盐 过碳酸钠 马桶清洁剂 草酸清洁剂 除垢剂 散装柠檬酸 水垢水锈 五洁粉"

def light_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):
    h = random.randint(30, 70)  # 选择浅色系的色调
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random.randint(70, 100)) / 255.0)  # 亮度选择较高的值
    return "hsl({}, {}%, {}%)".format(h, s, l)


# 创建词云对象，使用更亮的颜色
wordcloud = WordCloud(font_path="lantym.ttf", 
                      background_color="rgba(255, 255, 255, 0)",
                      mode="RGBA",width=800,height=800,
                      color_func=light_color_func
                      ).generate(text)

# 绘制词云图
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=0)

# 显示图像
plt.show()

# 保存词云图到文件
wordcloud.to_file("wordcloud.png")
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
from matplotlib.animation import FuncAnimation, PillowWriter

# 文本数据
text = "清洁用品 爆炸盐 过碳酸钠 马桶清洁剂 草酸清洁剂 除垢剂 散装柠檬酸 水垢水锈 五洁粉"

# 创建词云对象
wordcloud = WordCloud(font_path="lantym.ttf", width=800, height=400, background_color="rgba(255, 255, 255, 0)",mode="RGBA").generate(text)

# 初始化空白图像
fig, ax = plt.subplots()
ax.set_axis_off()
im = ax.imshow(np.zeros((wordcloud.height, wordcloud.width, 4)))

# 动画更新函数
def update(i):
    # 生成词云图
    wordcloud.generate(text)
    # 更新图像数据
    im.set_array(wordcloud.to_array())
    return [im]

# 创建动画
ani = FuncAnimation(fig, update, frames=range(10), interval=1000)

# 保存动画为 GIF 文件
writer = PillowWriter(fps=1)
ani.save("wordcloud_animation.gif", writer=writer)

# 显示动画
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# 数据
categories = ['A', 'B', 'C', 'D', 'E']
values1 = np.array([5, 10, 15, 20, 25])
values2 = np.array([7, 12, 17, 22, 27])

# 初始化画布和子图
fig, ax = plt.subplots()
ax.set_xlim(0, len(categories))
ax.set_ylim(0, max(max(values1), max(values2)) + 5)

# 条形图对象
bars1 = ax.bar(np.arange(len(categories)) - 0.2, values1, width=0.4, color='blue', label='Group 1')
bars2 = ax.bar(np.arange(len(categories)) + 0.2, values2, width=0.4, color='orange', label='Group 2')

# 添加标签和标题
ax.set_xticks(np.arange(len(categories)))
ax.set_xticklabels(categories)
ax.legend()
ax.set_title('Dynamic Bar Chart')

# 更新函数
def update(i):
    # 模拟数据变化
    values1_new = values1 + np.random.randint(-3, 3, size=len(categories))
    values2_new = values2 + np.random.randint(-3, 3, size=len(categories))
    
    # 更新条形图数据
    for bar1, bar2, h1, h2 in zip(bars1, bars2, values1_new, values2_new):
        bar1.set_height(h1)
        bar2.set_height(h2)
    
    return bars1 + bars2

# 创建动画
ani = FuncAnimation(fig, update, frames=range(10), blit=True)

# 保存动画为 GIF 文件
writer = PillowWriter(fps=0.5)
ani.save("wordcloud_animation.gif", writer=writer)


# 显示动画
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# 数据
products = ['Product A', 'Product B']
unit_prices = [50, 70]

# 初始化画布和子图
fig, ax = plt.subplots()
ax.set_xlim(0, max(unit_prices))
ax.set_ylim(0, len(products))

# 条形图对象
bars = ax.barh(products, [0]*len(products))

# 添加标签和标题
ax.set_xlabel('Unit Price')
ax.set_title('Dynamic Bar Chart of Unit Prices')

# 更新函数
def update(i):
    # 更新条形图数据
    for j, bar in enumerate(bars):
        bar.set_width(i / 10 * unit_prices[j])  # 每次增加10%
    
    return bars

# 创建动画
ani = FuncAnimation(fig, update, frames=range(11), blit=True)

# 保存动画为 GIF 文件
writer = PillowWriter(fps=0.5)
ani.save("wordcloud_animation.gif", writer=writer)

# 显示动画
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
from matplotlib.font_manager import FontProperties

# 数据
products = ['水池清洁剂（500ml）', '五洁粉（500g）']
unit_prices = [30, 6]

# 初始化画布和子图
fig, ax = plt.subplots()
ax.set_ylim(0, max(unit_prices))
ax.set_xlim(-0.5, len(products) - 0.5)

# 不同产品对应的颜色
colors = ['blue', 'red']

# 条形图对象
bars = ax.bar(np.arange(len(products)), [0]*len(products), color=colors)

# 创建 FontProperties 对象，指定字体文件路径
font_prop = FontProperties(fname='lantym.ttf')

# 添加标签和标题
ax.set_ylabel('价格(元)', fontproperties=font_prop)
ax.set_title('水池清洁单价对比', fontproperties=font_prop)

# 更新函数
def update(i):
    # 更新条形图数据
    for j, bar in enumerate(bars):
        bar.set_height(i / 10 * unit_prices[j])  # 每次增加10%
    
    return bars

# 创建动画
ani = FuncAnimation(fig, update, frames=range(11), blit=True, interval=100)

# 设置横轴刻度标签为中文
ax.set_xticks(np.arange(len(products)))
ax.set_xticklabels(products, fontproperties=font_prop)  # 使用中文字体

# 显示动画
plt.show()



# 保存动画为 GIF 文件
writer = PillowWriter(fps=10)
ani.save("cleaning.gif", writer=writer)
```
