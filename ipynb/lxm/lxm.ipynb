```python
# -*- coding: UTF-8 -*-
import jieba
from os import path  #用来获取文档的路径

#词云
from PIL import Image
import numpy as  np
import matplotlib.pyplot as plt
#词云生成工具
from wordcloud import WordCloud,ImageColorGenerator
#需要对中文进行处理
import matplotlib.font_manager as fm
```

```python
# 读取所有词
from openpyxl import load_workbook

wb = load_workbook('lxm.xlsx')
sheets = wb.worksheets   # 获取当前所有的sheet

print(sheets)
sheet1 = sheets[0]

# 获取第一列所有数据
col1 = []
for col in sheet1['A']:
    col1.append(col.value)
# print(col1)

text = ""
for key in col1:
    text = text + key + " "
# len(text)
text = text.replace('《','')
text = text.replace('》','')
text = text.replace('“','')
text = text.replace('”','')
text = text.replace('—','')
text = text.replace('·','')
text = text.replace('(','')
text = text.replace(')','')


print(text)
```

```python
#背景图
bg=np.array(Image.open("map2.jpeg"))
word = WordCloud(
    font_path="font.TTF",
    mask=bg,  # 设置背景图片
    background_color='white',  # 设置背景颜色
    max_font_size=50,  # 设置字体最大值
    min_font_size=5,
    max_words=3000,  # 设置最大显示字数
    stopwords={' ','《','》',''}  # 设置停用词，停用词则不在词云图中表示
     ).generate(text)
image = word.to_image()
image.show()

#保存云图
word.to_file("res.png")
```
